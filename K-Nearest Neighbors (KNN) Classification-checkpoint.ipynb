```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_excel(r"C:\Users\ABC\Downloads\MNIST_Digits.xlsx")
df
```

```python
X, y = df.drop("label", axis=1), df["label"]
```

```python
scaler = MinMaxScaler()        
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.25, random_state=42, stratify=y
)
```

```python
for k in [1, 3, 5, 7, 9]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    preds = knn.predict(X_test)
    acc  = accuracy_score(y_test, preds)
    print(f"k={k:<2}  accuracy={acc:.4f}")
```

```python
best_k = 3
knn_best = KNeighborsClassifier(n_neighbors=best_k).fit(X_train, y_train)
cm = confusion_matrix(y_test, knn_best.predict(X_test))
ConfusionMatrixDisplay(cm, display_labels=range(10)).plot(cmap="Blues")
plt.title(f"Kâ€‘NN (k={best_k}) â€“ confusion matrix")
plt.show()
```

```python
from sklearn.decomposition import PCA
import numpy as np

pca = PCA(n_components=2, random_state=42)
X2d = pca.fit_transform(X_scaled)

# Train on 2â€‘D projection just for plotting
knn_2d = KNeighborsClassifier(n_neighbors=3).fit(X2d, y)

# Meshâ€‘grid
xx, yy = np.meshgrid(
    np.linspace(X2d[:,0].min()-1, X2d[:,0].max()+1, 400),
    np.linspace(X2d[:,1].min()-1, X2d[:,1].max()+1, 400)
)
Z = knn_2d.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.figure()
plt.contourf(xx, yy, Z, alpha=.2)
scatter = plt.scatter(X2d[:,0], X2d[:,1], c=y, s=8, edgecolor="k")
plt.title("Kâ€‘NN decision regions (PCAâ€‘reduced)")
plt.xlabel("PCÂ 1"); plt.ylabel("PCÂ 2")
plt.show()
```
